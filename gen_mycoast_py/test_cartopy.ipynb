```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
from matplotlib.patches import PathPatch
import cartopy.crs as ccrs
import cartopy.io.shapereader as shpreader
from netCDF4 import Dataset

%matplotlib inline
```

```python
shp_filename='mycoast.shp'
mlon_c = 0.0
```

```python
map_proj = ccrs.PlateCarree(central_longitude = mlon_c)

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1, projection=map_proj)

gl = ax.gridlines(draw_labels=True)
gl.xlabels_top = False
gl.ylabels_right = False

all_info=[]
all_geom=[]

mycoast = shpreader.Reader(shp_filename)
for info, geom in zip(mycoast.records(), mycoast.geometries()):
    all_info.append(info)
    all_geom.append(geom)
    
ax.add_geometries(all_geom, crs=map_proj, facecolor='silver', edgecolor='k', linewidths=1.0)
```
